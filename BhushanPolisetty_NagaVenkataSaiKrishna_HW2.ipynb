```python
batting_stats['lgID'].fillna('NA',inplace=True)
```

```python
batting_stats.fillna(0,inplace=True)
```

```python
batting_stats['PA']=batting_stats['AB']+batting_stats['BB']+batting_stats['HBP']+batting_stats['SH']+batting_stats['SF']
```

```python
batting_stats['OBP']=(batting_stats['H']+batting_stats['BB']+batting_stats['HBP'])/(batting_stats['AB']+batting_stats['BB']+batting_stats['HBP']+batting_stats['SF'])
```

```python
batting_stats['BA']=(batting_stats['H'])/(batting_stats['AB'])
```

## 3)How many players have hit 40 or more HRs in one single season? (Number only)

```python
homeRuns=batting_stats.groupby(['playerID','yearID']).sum().reset_index()
```

```python
homeRuns.head()
```

```python
#players who hit 40 or more HRs in one single season
len(homeRuns[homeRuns.HR>=40][['playerID']].drop_duplicates())
```

## 4)How many players have hit 600 or more HRs for their career? (Dataframe with name and player first and last name only and amount of HRs)

```python
homeruns_career=batting_stats.groupby(['playerID']).sum().reset_index()
```

```python
homeruns_career.head()
```

```python
pd.merge(batting_stats[['playerID','nameFirst','nameLast']],
         homeruns_career[homeruns_career.HR>=600][['playerID','HR']],
         left_on='playerID',
         right_on='playerID',
         how='inner').drop_duplicates()
```

## 25)Who are the five youngest players to hit 300 or more HRs for their career? (Dataframe, first name, last name, season they eclipsed more than 300 HRs)

```python
homeruns_final=batting_stats.sort_values(['playerID','yearID'],ascending=True)
```

```python
homeruns_final['cumsum'] = homeruns_final.groupby(['playerID'])['HR'].cumsum()
homeruns_final['Age']=(homeruns_final['yearID']-homeruns_final['birthYear'])
```

```python
pd.merge(batting_stats[['playerID','nameFirst','nameLast']],
         homeruns_final[['playerID','yearID','cumsum','Age']][homeruns_final['cumsum']>=300].sort_values(['Age','yearID']).head(),
         left_on='playerID',
         right_on='playerID',
         how='inner').drop_duplicates().sort_values(['Age','yearID'])
```
